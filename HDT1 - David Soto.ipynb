#### Importamos el CSV con el uso de la libreria Pandas y revisamos el contenido de sus primeras 10 filas a modo de realizar una análisis exploratorio de los datos

```python
dataFrame = pd.read_csv("Admission_Predict.csv")
dataFrame.head(10)
```

#### Se separa el set de datos por columnas para tener los datos separados por variables

```python
serial_no = dataFrame['Serial No.']
gre_score = dataFrame['GRE Score']
toefl_score = dataFrame['TOEFL Score']
university_rating = dataFrame['University Rating']
sop = dataFrame['SOP']
lor = dataFrame['LOR ']
cgpa = dataFrame['CGPA']
research = dataFrame['Research']
chance_of_admit = dataFrame['Chance of Admit ']
```
